```python
# import BeautifulSoup and Requests library
from bs4 import BeautifulSoup
import requests
```

```python
#import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time

#import regex
import re
import pandas as pd
import numpy as np
```

```python
URL = "https://hk.centanet.com/findproperty/en/list/transaction?q=5C8BuPxhTEBERkEOTF6Gg&gclid=Cj0KCQjwvLOTBhCJARIsACVldV3U1jZYOof3e8nz6ZKUdlbXe_yMcm7S7e9lEFgV3bvSuiXxkXHPN3kaAmiKEALw_wcB"
driver = webdriver.Chrome('./chromedriver')
driver.get(URL)
subhtml = driver.page_source 
soup = BeautifulSoup(subhtml, "html.parser")
```

```python
#Page 1
for i in soup.find_all("div", class_="right-block row-cxt"): 
    print(i.find("span", class_="title-lg").text)
    print(i.find("span", class_="adress tag-adress").text)
    print(re.search('(\d,)*\d+\sft²',i.find("div",class_="num").find("span", class_="hidden-xs-only").text).group())
    #print(re.search('\$\d+',i.find("div",class_="area-price").text).group())
    print(re.search('\d+\-\d+\-\d+',i.find(class_="date").text).group())
    print(re.search('(\d+,)*\d+',i.find("div",class_="price").text).group())
    print('---------------\n')
```

```python
def extract():
    try: name = i.find("span", class_="title-lg").text
    except: name = "NaN"
    try: address = i.find("span", class_="adress tag-adress").text
    except: address = "NaN"  
    try: SA = re.search('(\d,)*\d+\sft²',i.find("div",class_="num").find("span", class_="hidden-xs-only").text).group()
    except: SA = "NaN" 
    #try: areaprice = re.search('\$\d+',i.find("div",class_="area-price").text).group()
    #except: areaprice = "NaN"
    try: date = re.search('\d+\-\d+\-\d+',i.find("div", class_="date").text).group()
    except: date = "NaN"
    try: rent = re.search('(\d+,)*\d+',i.find("div", class_="price").text).group()
    except: rent = "NaN"
        
    return (name, address, SA, date, rent)
```

```python
x = []
for i in soup.find_all("div", class_="right-block row-cxt"): 
    x.append(extract())
```

```python
trans_record = []

while True:
    #print(soup.find("div", class_="right-block row-cxt").find("span", class_="title-lg").text)
    for i in soup.find_all("div", class_="right-block row-cxt"): 
        trans_record.append(extract())
    
    try: 
        element = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.CLASS_NAME,"btn-next")))
        element.click()
    except:
        break
        
    try:
        WebDriverWait(driver,2) .until(EC.presence_of_element_located((By.CLASS_NAME,"right-block row-cxt")))
    except: print("-------")

    print(driver.current_url) #for checking
    subhtml = driver.page_source
    soup = BeautifulSoup(subhtml, "html.parser")
    
    
#print(soup.find("div", class_="right-block row-cxt").find("span", class_="title-lg").text)
for i in soup.find_all("div", class_="right-block row-cxt"): 
    trans_record.append(extract())
```
